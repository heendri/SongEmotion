```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import os
import glob
import csv
%matplotlib inline
```

```python
data = []
directory = os.listdir('raw2/')
for dirs in directory:
    for file in glob.glob("raw2/"+dirs+"/*.txt"):
        with open(file) as open_f:
            tulis = []
            tag_isi = re.findall(r'(?:<h2>.*?<div style="clear:both;">)', open_f.read(), flags=re.DOTALL)
            for isi in tag_isi:
                tag_judul = re.compile(r'<h2>(.*)</h2>')
                judul = tag_judul.findall(isi)
                for i in judul:
                    #print("Judul: "+i+"\n")
                    tulis.append(i)
                tag_isi = re.compile(r'<p>(.*)</p>')
                lirik_ = tag_isi.findall(isi)
                lirik = ""
                for i in lirik_:
                    tmp = re.sub(r'<ins .*?/ins>', '', i)
                    tmp = re.sub(r'<script.*?/script>', '', tmp)
                    tmp = re.sub(r'<.*?>','',tmp)
                    #print(tmp)
                    lirik = lirik + " " + tmp
                #print(lirik)
                tulis.append(lirik)
        data.append(tulis)
dataa = pd.DataFrame(data, columns=['Judul','Lirik'])
#dataa
#dataset = dataa['Judul'].isnull()
dataset = dataa[(dataa['Judul'].isnull()==False) | (dataa['Lirik'].isnull()==False)]
dataset.to_csv('output2.csv')
dataset
```

```python
data = []
directory = os.listdir('raw2/')
for dirs in directory:
    for file in glob.glob("raw2/"+dirs+"/*.txt"):
        with open(file) as open_f:
            tulis = []
            tag_isi = re.findall(r'(?:<h1.*?<div style="clear:both;">)', open_f.read(), flags=re.DOTALL)
            for isi in tag_isi:
                tag_judul = re.compile(r'<h1 .*>(.*)</h1>')
                judul = tag_judul.findall(isi)
                for i in judul:
                    tmp = i.split(" - ")
                    artist = re.sub(r'Lirik','',tmp[0])
                    title = tmp[1]
                    #print("Judul: "+title+", Artist: "+artist+"\n")
                    tulis.append(title)
                    tulis.append(artist)
                tag_gambar = re.compile(r'<img .*src="(.*)" class=".*>')
                tag_isi = re.compile(r'<p>(.*)</p>')
                lirik_ = tag_isi.findall(isi)
                gambar = tag_gambar.findall(isi)
                for i in gambar:
                    #print(i+"\n\n")
                    gambar = i
                lirik = ""
                for i in lirik_:
                    tmp = re.sub(r'<ins .*?/ins>', '', i)
                    tmp = re.sub(r'<script.*?/script>', '', tmp)
                    tmp = re.sub(r'<.*?>','',tmp)
                    #print(tmp)
                    lirik = lirik + " " + tmp
                #print(lirik)
                tulis.append(lirik)
                tulis.append(gambar)
        data.append(tulis)
dataa = pd.DataFrame(data, columns=['Judul','Artist','Lirik','Gambar'])
#dataa
dataset = dataa[(dataa['Judul'].isnull()==False) | (dataa['Lirik'].isnull()==False) | (dataa['Artist'].isnull()==False) | (dataa['Gambar'].isnull==False)]
dataset.to_csv('output2.csv')
dataset
```
